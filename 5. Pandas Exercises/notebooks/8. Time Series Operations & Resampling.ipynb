```python
import pandas as pd
import numpy as np
# Relative path from notebooks/ to data/
file_path = "../data/google_stock_data.xlsx"

# Read the Excel file
price = pd.read_excel(file_path)

# Round all numerical columns to 2 decimal places
price = price.round(2)

# update the date format to YYYY-MM-DD
price['Date'] = pd.to_datetime(price['Date']).dt.strftime('%Y-%m-%d')

# Rename columns
price.rename(columns={'Date': 'date', 'Close_GOOG': 'close', 'High_GOOG': 'high'
                   ,'Low_GOOG': 'low', 'Open_GOOG': 'open'
                   ,'Volume_GOOG': 'volume'}, inplace=True)

price['date'] = pd.to_datetime(price['date'])
price = price.sort_values('date')

price.set_index('date', inplace=True)
```

1. Calculate the monthly average close price.

```python
price['monthly_average_close'] = price.resample('M')['close'].transform('mean')
price
```

2. Find the total trading volume per month.

```python
price['monthly_total_trading_volume'] = price.resample('M')['volume'].transform('sum')
price
```

3. For each month, find the maximum high price

```python
price['monthly_maximum_high_price'] = price.resample('M')['high'].transform('max')
price
```

4. Create weekly OHLC data from daily prices

```python
price['close'].resample('W').ohlc()
```

5. Calculate a 7-day rolling average of close price.

```python
price['7_day_rolling_avg_close'] = price['close'].rolling(7).mean()
price
```

6. Calculate 7-day rolling standard deviation of close price

```python
price['7_day_rolling_std_close'] = price['close'].rolling(7).std()
price
```

7. Compute daily price change in close price

```python
price['daily_range_close'] = price['close'].diff()
price
```

8. Compute daily percentage return

```python
price['daily_percent_return'] = price['daily_range_close'].pct_change()
price
```

9. Create a column that shows whether today’s close is higher than yesterday.

```python
price['close_high'] = price['close'] > price['close'].shift(1)
price
```

10. For each month, find the date with the highest close price

```python
price['monthly_highest_close'] = price.resample('M')['close'].transform('max')
price
```
