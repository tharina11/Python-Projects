```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
import pandas as pd
```

```python
# Load dataset
iris = datasets.load_iris() 

# Display data in a dataframe
df= pd.DataFrame(data= np.c_[iris['data'], iris['target']], columns= iris['feature_names'] + ['target'])
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
df.head()
```

```python
# Number of iris species in the dataset
df['species'].unique()
```

```python
# separate data into X and y variables
X = iris.data[:, :4] # All four measurements of the flowers
y = iris.target # Flower names 
```

```python
# Create a 3D space to plot data
fig = plt.figure(figsize=(8, 6)) # The parent figure 
ax = Axes3D(fig) # Create a 3D space object using the parent figure 

# Transform iris flower data into three principal components
X_reduced = decomposition.PCA(n_components=3).fit_transform(iris.data)

# Plot data in a scatterplot and color base on the target class (flower names)
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=y, edgecolor='k', s=40)
# Title and axes labels
ax.set_title("First three Principal Components of Iris Flowers")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])

plt.show()
```
