### What is Recursion?
Recursion is calling a function from itself. 

```python
def greet():
    print("Hello")
```

```python
greet()
```

The function simply prints the word hello. If we need to print this muptiple times, one way to do it is calling it iteratively using a loop. Recursion is another way to do this.

```python
def greet():
    print("Hello")
    greet()
```

```python
greet()
```

The code does not print 'Hello' infinitely, because a recursive function has a maximum recursive depth limit.

```python
import sys
print(sys.getrecursionlimit())
```

The recursion limit for the Python interpreter is 3000. Let's change it to 4000.

```python
sys.setrecursionlimit(4000)
print(sys.getrecursionlimit())
```

```python
i = 0

def greet():
    global i
    i+=1
    print("Hello", i)
    greet()
```

```python
greet()
```

The code printed Hello for 3971 times.
